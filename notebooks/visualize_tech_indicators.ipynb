````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#To display plots inline in the notebook.
%matplotlib inline 
````

````python
# Set pandas options to display all columns
pd.set_option('display.max_columns', None)
````

Load the Data. For the seek of this visualization we'll focus on the BTC data. 

````python
# Set the path to your data file
data_file = '../data/with_indicators/BTC_data.csv'

# Read the data into a DataFrame
df = pd.read_csv(data_file)

# Display the first few rows
df.head()
````

- Convert the 'Datetime' column to datetime format.
- Set the datetime column as the index for easier plotting.

````python
# Check if 'Datetime' or 'Date' column exists
if 'Datetime' in df.columns:
    df['Datetime'] = pd.to_datetime(df['Datetime'])
    df.set_index('Datetime', inplace=True)
elif 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
else:
    raise ValueError("No 'Datetime' or 'Date' column found in the data.")

# Display the DataFrame info
df.info()
````

### Plot the technical indicators

````python
# FOCUS THE ANALYSIS ON A SPECIFIC DATE RANGE

# Define the date range
start_date = '2024-01-01'
end_date = '2024-03-01'

# Subset the DataFrame
df = df.loc[start_date:end_date]
````

Momentum (MOM)

````python
# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(15, 6))

# Plot the Close price
ax1.set_title('Bitcoin Close Price and Momentum (MOM)', fontsize=16)
ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('Close Price (USD)', fontsize=14, color='blue')
ax1.plot(df.index, df['Close'], label='Close Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a twin axis to plot MOM
ax2 = ax1.twinx()
ax2.set_ylabel('Momentum (MOM)', fontsize=14, color='red')
ax2.plot(df.index, df['MOM'], label='Momentum (MOM)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
````

Plotting Chande Momentum Oscillator (CMO).
Chande Momentum Oscillator (CMO) measures the momentum on both upward and downward movements.

Interpretation Tips:

- CMO Range: Oscillates between -100 and +100.
- Overbought Zone: CMO > +50 (possible sell signal).
- Oversold Zone: CMO < -50 (possible buy signal).
- Look for crossovers of the CMO with these levels.

````python
# Choose a CMO period to plot, e.g., 14
cmo_period = 14
cmo_column = f'CMO_{cmo_period}'

# Check if the CMO column exists
if cmo_column not in df.columns:
    raise ValueError(f"{cmo_column} not found in the DataFrame.")

# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(15, 6))

# Plot the Close price
ax1.set_title(f'Bitcoin Close Price and Chande Momentum Oscillator (CMO_{cmo_period})', fontsize=16)
ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('Close Price (USD)', fontsize=14, color='blue')
ax1.plot(df.index, df['Close'], label='Close Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a twin axis to plot CMO
ax2 = ax1.twinx()
ax2.set_ylabel(f'CMO_{cmo_period}', fontsize=14, color='green')
ax2.plot(df.index, df[cmo_column], label=f'CMO_{cmo_period}', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Add horizontal lines at +50 and -50 levels
ax2.axhline(50, color='grey', linestyle='--', linewidth=1)
ax2.axhline(-50, color='grey', linestyle='--', linewidth=1)
ax2.axhline(0, color='grey', linestyle='-', linewidth=1)

# Add legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
````

Plotting Ultimate Oscillator (UO)
Ultimate Oscillator (UO) combines short-term, medium-term, and long-term price action into one oscillator.

Interpretation Tips:

- UO Range: Oscillates between 0 and 100.
- Overbought Zone: UO > 70 (possible sell signal).
- Oversold Zone: UO < 30 (possible buy signal).
- Look for divergences and crossovers with these levels.

````python
# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(15, 6))

# Plot the Close price
ax1.set_title('Bitcoin Close Price and Ultimate Oscillator (UO)', fontsize=16)
ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('Close Price (USD)', fontsize=14, color='blue')
ax1.plot(df.index, df['Close'], label='Close Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a twin axis to plot UO
ax2 = ax1.twinx()
ax2.set_ylabel('Ultimate Oscillator (UO)', fontsize=14, color='purple')
ax2.plot(df.index, df['UO'], label='Ultimate Oscillator (UO)', color='purple')
ax2.tick_params(axis='y', labelcolor='purple')

# Add horizontal lines at 30 and 70 levels
ax2.axhline(30, color='grey', linestyle='--', linewidth=1)
ax2.axhline(70, color='grey', linestyle='--', linewidth=1)
ax2.axhline(50, color='grey', linestyle='-', linewidth=1)

# Add legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
````

````python
# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(15, 6))

# Plot the Close price
ax1.set_title('Bitcoin Close Price and Ultimate Oscillator (UO) with Anomalies', fontsize=16)
ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('Close Price (USD)', fontsize=14, color='blue')
ax1.plot(df.index, df['Close'], label='Close Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Highlight anomalies
anomaly_df = df[df['Anomaly'] != 0]
ax1.scatter(anomaly_df.index, anomaly_df['Close'], color='orange', label='Anomalies', marker='x')

# Create a twin axis to plot UO
ax2 = ax1.twinx()
ax2.set_ylabel('Ultimate Oscillator (UO)', fontsize=14, color='purple')
ax2.plot(df.index, df['UO'], label='Ultimate Oscillator (UO)', color='purple')
ax2.tick_params(axis='y', labelcolor='purple')

# Add horizontal lines at 30 and 70 levels
ax2.axhline(30, color='grey', linestyle='--', linewidth=1)
ax2.axhline(70, color='grey', linestyle='--', linewidth=1)
ax2.axhline(50, color='grey', linestyle='-', linewidth=1)

# Add legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
````

Correlation HeatMap

````python
# Select relevant columns
columns_of_interest = ['Close', 'MOM', cmo_column, 'UO']
df_corr = df[columns_of_interest].dropna()

# Compute the correlation matrix
corr_matrix = df_corr.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Technical Indicators and Close Price')
plt.show()
````
